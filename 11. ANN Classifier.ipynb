```python
from sklearn.preprocessing import OneHotEncoder
oh=OneHotEncoder(categorical_features=[1])
x=oh.fit_transform(x).toarray()  
```

```python
x=x[:,1:]
```

```python
from sklearn.cross_validation import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

# Standard Scaling 

<img src="https://i.stack.imgur.com/PZgJ2.pngg" width=500 height=500 />


```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()

x_train =sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
keras.__version__
```

```python
model=Sequential()
```

```python
x_train.shape
```

```python
model.add(Dense(input_dim=11,init="random_uniform",activation='relu',output_dim=15))#input layer
```

```python
model.add(Dense(output_dim=8,init="random_uniform",activation='relu'))#hidden layer
```

```python
model.add(Dense(output_dim=1,init='random_uniform',activation='sigmoid'))#output layer
```

# since it is a binary classification WE use binary cross entropy.if it is categorical then we ue categorical entropy

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])#adam=batch gradent descent
```

```python
model.fit(x_train,y_train,epochs=50,batch_size=32)#epochs no of iteration 
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred=(y_pred>0.5)
```
